```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Exploratory Data Analysis

## Gender

```python
ax=sns.countplot(x='Gender', data=df)
plt.show()
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Purchase'].sum().sort_values(by='Purchase',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Purchase'].sum().sort_values(by='Purchase',ascending=False)
```

```python
sns.barplot(x='Gender',y='Purchase',data=sales_gen)
```

### From the above graphs we can see that most of the buyers are males and even the purchasing power of males is greater than females.

## Age

```python
sns.countplot(data=df,x='Age')
```

```python
ax=sns.countplot(data=df,x='Age', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age'],as_index=False)['Purchase'].sum().sort_values (by='Purchase',ascending=False)
```

```python
sales_age=df.groupby(['Age'],as_index=False)['Purchase'].sum().sort_values (by='Purchase',ascending=False)
sns.barplot(x='Age',y='Purchase',data=sales_age)
```

### From the above graphs we can see that most of the buyers are males of age group 25-36.

## State

```python
sale_state=df.groupby(['City_Category'],as_index=False)['Purchase'].sum().sort_values (by='Purchase',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='City_Category',y='Purchase',data=sale_state)

ax=sns.countplot(data=df,x='City_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_counts = df['Gender'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values
explode = (0.1,0)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode)
plt.axis('equal')
plt.show()
```

### From above graphs we can see that most of the purchase & total sales/amount are from B category cities.

## Occupation

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Purchase'].sum().sort_values(by='Purchase', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Purchase')

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

### From above graphs we can see that most of the buyers are working in 4, 0 and 7 sector.

## Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Purchase'].sum().sort_values(by='Purchase', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Purchase', hue='Gender')
```

### From above graphs we can see that most of the buyers are married (men) and they have high purchasing power.

## Product Category

```python
sales_state = df.groupby(['Product_Category_1'], as_index=False)['Purchase'].sum().sort_values(by='Purchase', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category_1',y= 'Purchase')

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category_1')

for bars in ax.containers:
    ax.bar_label(bars)
```

### From above graphs we can see that most of the sold products are from 1, 5 and 8 category.

## Product ID

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Purchase'].sum().sort_values(by='Purchase', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Purchase')

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Purchase'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

# Conclusion

### Married men age group 26-35 yrs from B, C and A category cities working in 4, 0 and sectors are more likely to buy products from 1, 5 and 8 category.
